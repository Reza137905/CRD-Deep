```
import os
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
import shap
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import LabelEncoder, StandardScaler
import lightgbm as lgb
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score
from cryptography.fernet import Fernet
import warnings
from collections import deque
import random
import torch.nn.functional as F
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier, AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
import xgboost as xgb
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import euclidean_distances

# نادیده گرفتن هشدارهای مربوط به کتابخانه‌ها
warnings.filterwarnings("ignore", category=UserWarning, module='keras')
warnings.filterwarnings("ignore", category=UserWarning, module='shap')
warnings.filterwarnings("ignore", category=UserWarning, module='sklearn')
warnings.filterwarnings("ignore", category=UserWarning, module='torch_geometric')

# ==========================================================
# نصب کتابخانه‌های مورد نیاز
# ==========================================================
def install_libraries():
    print("در حال نصب کتابخانه‌های مورد نیاز...")
    try:
        os.system('pip install gdown==4.4.0 --quiet')
        os.system('pip install lightgbm==3.3.5 --quiet')
        os.system('pip install scikit-learn==1.2.2 --quiet')
        os.system('pip install shap==0.42.1 --quiet')
        os.system('pip install cryptography==41.0.3 --quiet')
        os.system('pip install xgboost --quiet')

        # نصب torch_geometric به صورت خاص
        if not os.path.exists('torch_geometric_installed'):
            print("در حال نصب torch_geometric...")
            os.system('pip install torch-scatter torch-sparse torch-cluster torch-spline-conv torch-geometric -f https://data.pyg.org/whl/torch-2.0.0+cu118.html --quiet')
            with open('torch_geometric_installed', 'w') as f:
                f.write('Installed')

        try:
            from torch_geometric.data import Data
            print("torch_geometric با موفقیت نصب و وارد شد.")
        except ImportError:
            print("خطا در نصب torch_geometric. نصب دستی را امتحان کنید.")
            print("برای نصب دستی، از دستور زیر استفاده کنید:")
            print("pip install torch-scatter torch-sparse torch-cluster torch-spline-conv torch-geometric -f https://data.pyg.org/whl/torch-2.0.0+cu118.html")
    except Exception as e:
        print(f"خطا در نصب کتابخانه‌ها: {e}")

# ===============================
# دانلود فایل CSV از Google Drive
# ===============================
def download_data():
    try:
        import gdown
        file_id = "1YMO2gog6AGiZhlgLCtgHb8Yr3QPxoZ0Q"
        output_file = "telecocustomers.csv"
        if not os.path.exists(output_file):
            print("در حال دانلود فایل داده‌ها...")
            gdown.download(f"https://drive.google.com/uc?id={file_id}", output_file, quiet=False)
        return output_file
    except Exception as e:
        print(f"خطا در دانلود فایل: {e}. مطمئن شوید gdown نصب است.")
        return None

# ===============================
# تعریف معماری‌های شبکه عصبی
# ===============================
class ResNetForTabular(nn.Module):
    def __init__(self, input_dim, num_classes=1):
        super(ResNetForTabular, self).__init__()
        self.fc_in = nn.Linear(input_dim, 128)
        self.block1 = nn.Sequential(nn.Linear(128, 128), nn.ReLU(), nn.Linear(128, 128))
        self.block2 = nn.Sequential(nn.Linear(128, 128), nn.ReLU(), nn.Linear(128, 128))
        self.fc_out = nn.Linear(128, num_classes)
        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()
    def forward(self, x):
        x = self.relu(self.fc_in(x))
        x = x + self.block1(x)
        x = x + self.block2(x)
        x = self.fc_out(x)
        return self.sigmoid(x)

class EfficientNetLike(nn.Module):
    def __init__(self, input_dim, num_classes=1):
        super(EfficientNetLike, self).__init__()
        self.layers = nn.Sequential(
            nn.Linear(input_dim, 256), nn.ReLU(),
            nn.Linear(256, 128), nn.ReLU(),
            nn.Linear(128, 64), nn.ReLU(),
            nn.Linear(64, num_classes), nn.Sigmoid()
        )
    def forward(self, x):
        return self.layers(x)

class SharedMTLModel(nn.Module):
    def __init__(self, input_dim):
        super(SharedMTLModel, self).__init__()
        self.shared_layers = nn.Sequential(
            nn.Linear(input_dim, 256), nn.ReLU(), nn.Dropout(0.3),
            nn.Linear(256, 128), nn.ReLU(), nn.Dropout(0.3)
        )
        self.churn_head = nn.Sequential(
            nn.Linear(128, 64), nn.ReLU(), nn.Linear(64, 1), nn.Sigmoid()
        )
        self.cltv_head = nn.Sequential(
            nn.Linear(128, 64), nn.ReLU(), nn.Linear(64, 1)
        )
    def forward(self, x):
        shared_features = self.shared_layers(x)
        churn_output = self.churn_head(shared_features)
        cltv_output = self.cltv_head(shared_features)
        return churn_output, cltv_output

class CNNLSTM(nn.Module):
    def __init__(self, input_dim, num_classes=1):
        super(CNNLSTM, self).__init__()
        self.conv1d = nn.Conv1d(in_channels=1, out_channels=64, kernel_size=3, padding=1)
        self.relu = nn.ReLU()
        self.pool1d = nn.MaxPool1d(kernel_size=2)
        lstm_input_dim = 64 * (input_dim // 2)
        self.lstm = nn.LSTM(input_size=lstm_input_dim, hidden_size=128, batch_first=True)
        self.fc = nn.Linear(128, num_classes)
        self.sigmoid = nn.Sigmoid()
    def forward(self, x):
        x = x.unsqueeze(1)
        x = self.conv1d(x)
        x = self.relu(x)
        x = self.pool1d(x)
        x = x.view(x.size(0), 1, -1)
        _, (hn, _) = self.lstm(x)
        out = self.fc(hn.squeeze(0))
        return self.sigmoid(out)

class XceptionLike(nn.Module):
    def __init__(self, input_dim, num_classes=1):
        super(XceptionLike, self).__init__()
        grid_dim = int(np.ceil(np.sqrt(input_dim)))
        self.input_dim = input_dim
        self.grid_dim = grid_dim
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=32, kernel_size=3, padding=1)
        self.sep_conv1 = nn.Sequential(
            nn.Conv2d(32, 32, kernel_size=3, groups=32, padding=1),
            nn.Conv2d(32, 64, kernel_size=1)
        )
        self.relu = nn.ReLU()
        self.avg_pool = nn.AdaptiveAvgPool2d((1, 1))
        self.fc = nn.Linear(64, num_classes)
        self.sigmoid = nn.Sigmoid()
    def forward(self, x):
        x_padded = torch.zeros(x.size(0), self.grid_dim * self.grid_dim).to(x.device)
        x_padded[:, :self.input_dim] = x
        x_2d = x_padded.view(-1, 1, self.grid_dim, self.grid_dim)
        out = self.relu(self.conv1(x_2d))
        out = self.relu(self.sep_conv1(out))
        out = self.avg_pool(out)
        out = out.view(out.size(0), -1)
        out = self.fc(out)
        return self.sigmoid(out)

class GNNModel(nn.Module):
    def __init__(self, input_dim, num_classes=1):
        super(GNNModel, self).__init__()
        self.conv1 = GCNConv(input_dim, 128)
        self.conv2 = GCNConv(128, 64)
        self.fc = nn.Linear(64, num_classes)
    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, p=0.5, training=self.training)
        x = self.conv2(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, p=0.5, training=self.training)
        return torch.sigmoid(self.fc(x))

class GCRNN(nn.Module):
    def __init__(self, input_dim, num_classes=1):
        super(GCRNN, self).__init__()
        self.conv1 = GCNConv(input_dim, 128)
        self.rnn = nn.GRU(128, 64, batch_first=True)
        self.fc = nn.Linear(64, num_classes)
    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = x.unsqueeze(0)
        x, _ = self.rnn(x)
        x = x.squeeze(0)
        return torch.sigmoid(self.fc(x))

class RNNModel(nn.Module):
    def __init__(self, input_dim, hidden_dim=64, num_layers=1, num_classes=1):
        super(RNNModel, self).__init__()
        self.rnn = nn.RNN(input_dim, hidden_dim, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_dim, num_classes)
    def forward(self, x):
        x = x.unsqueeze(1)
        _, hn = self.rnn(x)
        out = self.fc(hn.squeeze(0))
        return torch.sigmoid(out)

class BiLSTMCNN(nn.Module):
    def __init__(self, input_dim, hidden_dim=64, num_layers=1, num_classes=1):
        super(BiLSTMCNN, self).__init__()
        self.conv1d = nn.Conv1d(in_channels=1, out_channels=64, kernel_size=3, padding=1)
        self.pool1d = nn.MaxPool1d(kernel_size=2)
        cnn_output_dim = 64 * (input_dim // 2)
        self.bilstm = nn.LSTM(cnn_output_dim, hidden_dim, num_layers, batch_first=True, bidirectional=True)
        self.fc = nn.Linear(hidden_dim * 2, num_classes)
    def forward(self, x):
        x = x.unsqueeze(1)
        x = self.conv1d(x)
        x = F.relu(x)
        x = self.pool1d(x)
        x = x.view(x.size(0), 1, -1)
        _, (hn, _) = self.bilstm(x)
        hn = torch.cat((hn[-2,:,:], hn[-1,:,:]), dim=1)
        out = self.fc(hn)
        return torch.sigmoid(out)

class DDQN(nn.Module):
    def __init__(self, state_dim, action_dim):
        super(DDQN, self).__init__()
        self.net = nn.Sequential(
            nn.Linear(state_dim, 128), nn.ReLU(),
            nn.Linear(128, 128), nn.ReLU(),
            nn.Linear(128, action_dim)
        )
    def forward(self, state):
        return self.net(state)

class DDQNAgent:
    def __init__(self, state_dim, action_dim, gamma=0.99, lr=0.001, device='cpu'):
        self.state_dim = state_dim
        self.action_dim = action_dim
        self.gamma = gamma
        self.device = device
        self.memory = deque(maxlen=2000)
        self.policy_net = DDQN(state_dim, action_dim).to(device)
        self.target_net = DDQN(state_dim, action_dim).to(device)
        self.target_net.load_state_dict(self.policy_net.state_dict())
        self.target_net.eval()
        self.optimizer = torch.optim.Adam(self.policy_net.parameters(), lr=lr)
        self.loss_fn = nn.MSELoss()

    def choose_action(self, state, epsilon):
        if random.random() < epsilon:
            return random.randrange(self.action_dim)
        with torch.no_grad():
            q_values = self.policy_net(state)
            return q_values.argmax().item()

    def learn(self, batch_size):
        if len(self.memory) < batch_size:
            return
        transitions = random.sample(self.memory, batch_size)
        states, actions, rewards, next_states, dones = zip(*transitions)
        states = torch.tensor(np.array(states), dtype=torch.float32).to(self.device)
        actions = torch.tensor(actions, dtype=torch.long).to(self.device)
        rewards = torch.tensor(rewards, dtype=torch.float32).to(self.device)
        next_states = torch.tensor(np.array(next_states), dtype=torch.float32).to(self.device)
        dones = torch.tensor(dones, dtype=torch.float32).to(self.device)

        q_values = self.policy_net(states).gather(1, actions.unsqueeze(1)).squeeze(1)

        next_q_values_target = self.target_net(next_states).detach()
        next_q_values_policy = self.policy_net(next_states).detach()
        max_q_values = next_q_values_target.gather(1, next_q_values_policy.argmax(dim=1, keepdim=True)).squeeze(1)
        expected_q_values = rewards + self.gamma * max_q_values * (1 - dones)

        loss = self.loss_fn(q_values, expected_q_values)
        self.optimizer.zero_grad()
        loss.backward()
        self.optimizer.step()

    def update_target_net(self):
        self.target_net.load_state_dict(self.policy_net.state_dict())

class Environment:
    def __init__(self, states, rewards):
        self.states = states
        self.rewards = rewards
        self.current_state_idx = 0
    def reset(self):
        self.current_state_idx = random.randrange(len(self.states))
        return self.states[self.current_state_idx]
    def step(self, action):
        reward = self.rewards[self.current_state_idx]
        done = True
        return self.states[self.current_state_idx], reward, done

# ===============================
# توابع آموزش و ارزیابی
# ===============================
def get_optimizer(model, optimizer_name, lr=0.001):
    if optimizer_name.lower() == 'adam':
        return torch.optim.Adam(model.parameters(), lr=lr)
    elif optimizer_name.lower() == 'sgd':
        return torch.optim.SGD(model.parameters(), lr=lr, momentum=0.9)
    elif optimizer_name.lower() == 'rmsprop':
        return torch.optim.RMSprop(model.parameters(), lr=lr)
    elif optimizer_name.lower() == 'adadelta':
        return torch.optim.Adadelta(model.parameters(), lr=1.0)
    else:
        raise ValueError(f"بهینه‌ساز '{optimizer_name}' پشتیبانی نمی‌شود.")

def train_model(model, train_loader, optimizer, criterion, epochs=50, device='cpu'):
    model.train()
    for epoch in range(epochs):
        total_loss = 0
        for inputs, targets in train_loader:
            inputs, targets = inputs.to(device), targets.to(device)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, targets.unsqueeze(1).float())
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
        if (epoch + 1) % 10 == 0:
            print(f'Epoch [{epoch+1}/{epochs}], Loss: {total_loss / len(train_loader):.4f}')

def train_mtl_model(model, train_loader, optimizer, criterion_churn, criterion_cltv, epochs=50, device='cpu'):
    model.train()
    for epoch in range(epochs):
        total_loss = 0
        for inputs, targets_churn, targets_cltv in train_loader:
            inputs, targets_churn, targets_cltv = inputs.to(device), targets_churn.to(device), targets_cltv.to(device)
            optimizer.zero_grad()
            churn_preds, cltv_preds = model(inputs)
            loss_churn = criterion_churn(churn_preds.squeeze(), targets_churn.float())
            loss_cltv = criterion_cltv(cltv_preds.squeeze(), targets_cltv.float())
            combined_loss = loss_churn + loss_cltv
            total_loss += combined_loss.item()
            combined_loss.backward()
            optimizer.step()
        if (epoch+1) % 10 == 0:
            print(f'Epoch [{epoch+1}/{epochs}], Combined Loss: {total_loss / len(train_loader):.4f}')

def train_ddqn(agent, env, num_episodes=50, batch_size=32):
    for episode in range(num_episodes):
        state = env.reset()
        done = False
        total_reward = 0
        epsilon = max(0.01, 0.5 - episode / 25)
        while not done:
            state_tensor = torch.tensor(state, dtype=torch.float32).to(agent.device)
            action = agent.choose_action(state_tensor, epsilon)
            next_state, reward, done = env.step(action)
            agent.memory.append((state, action, reward, next_state, done))
            state = next_state
            total_reward += reward
            if len(agent.memory) > batch_size:
                agent.learn(batch_size)
        if (episode + 1) % 10 == 0:
            agent.update_target_net()
            print(f"Episode {episode+1}/{num_episodes}, Total Reward: {total_reward:.2f}")
    print("آموزش DDQN کامل شد.")

def evaluate_all_models(models_dict, X_test, y_test, scaler, device):
    results = {}
    X_test_scaled = scaler.transform(X_test)
    test_loader = DataLoader(TensorDataset(torch.tensor(X_test_scaled, dtype=torch.float32), torch.tensor(y_test, dtype=torch.float32)), batch_size=64, shuffle=False)

    def evaluate_mtl_model(model, data_loader, device):
        model.eval()
        predictions = []
        with torch.no_grad():
            for inputs, _, _ in data_loader:
                inputs = inputs.to(device)
                churn_preds, _ = model(inputs)
                predictions.extend(churn_preds.cpu().numpy().flatten())
        return np.array(predictions)

    classic_models = {
        'LightGBM': models_dict.get('lgbm'),
        'XGBoost': models_dict.get('xgboost'),
        'DecisionTree': models_dict.get('decisiontree'),
        'RandomForest': models_dict.get('randomforest'),
        'Bagging': models_dict.get('bagging'),
        'AdaBoost': models_dict.get('adaboost'),
        'KNN': models_dict.get('knn'),
        'SVM': models_dict.get('svm'),
        'XGBoost-Optimized': models_dict.get('xgboost_optimized')
    }

    for name, model in classic_models.items():
        if model is None:
            continue
        if name in ['SVM', 'Bagging', 'AdaBoost', 'DecisionTree', 'RandomForest', 'KNN', 'LightGBM', 'XGBoost', 'XGBoost-Optimized']:
            if name == 'SVM':
                preds = model.decision_function(X_test_scaled)
            else:
                preds = model.predict_proba(X_test_scaled)[:, 1]
            results[name] = {
                'AUC': roc_auc_score(y_test, preds),
                'F1-Score': f1_score(y_test, (preds > 0.5).astype(int)),
                'Accuracy': accuracy_score(y_test, (preds > 0.5).astype(int))
            }

    for name, model in [('ResNet', models_dict.get('resnet')),
                         ('EfficientNet', models_dict.get('efficientnet')),
                         ('CNNLSTM', models_dict.get('cnnlstm')),
                         ('Xception', models_dict.get('xception')),
                         ('RNN', models_dict.get('rnn')),
                         ('BiLSTM-CNN', models_dict.get('bilstm-cnn'))]:
        if model is None:
            continue
        model.eval()
        with torch.no_grad():
            preds = model(torch.tensor(X_test_scaled, dtype=torch.float32).to(device)).cpu().numpy().flatten()
            results[name] = {
                'AUC': roc_auc_score(y_test, preds),
                'F1-Score': f1_score(y_test, (preds > 0.5).astype(int)),
                'Accuracy': accuracy_score(y_test, (preds > 0.5).astype(int))
            }

    if models_dict.get('mtl'):
        mtl_churn_preds = evaluate_mtl_model(models_dict['mtl'], DataLoader(TensorDataset(torch.tensor(X_test_scaled, dtype=torch.float32), torch.tensor(y_test, dtype=torch.float32), torch.tensor(np.zeros_like(y_test), dtype=torch.float32)), batch_size=64, shuffle=False), device=device)
        results['Shared-MTL (Churn)'] = {
            'AUC': roc_auc_score(y_test, mtl_churn_preds),
            'F1-Score': f1_score(y_test, (mtl_churn_preds > 0.5).astype(int)),
            'Accuracy': accuracy_score(y_test, (mtl_churn_preds > 0.5).astype(int))
        }

    # GNN & GCRNN
    if models_dict.get('gnn') and models_dict.get('gcrnn'):
        adj_matrix = euclidean_distances(X_test_scaled, X_test_scaled)
        edge_indices = np.where(adj_matrix < 2.0)
        edge_index = torch.tensor([edge_indices[0], edge_indices[1]], dtype=torch.long).to(device)
        graph_data_test = Data(x=torch.tensor(X_test_scaled, dtype=torch.float32).to(device), edge_index=edge_index)

        gnn_preds = models_dict['gnn'](graph_data_test).cpu().detach().numpy().flatten()
        results['GNN'] = {
            'AUC': roc_auc_score(y_test, gnn_preds),
            'F1-Score': f1_score(y_test, (gnn_preds > 0.5).astype(int)),
            'Accuracy': accuracy_score(y_test, (gnn_preds > 0.5).astype(int))
        }

        gcrnn_preds = models_dict['gcrnn'](graph_data_test).cpu().detach().numpy().flatten()
        results['GCRNN'] = {
            'AUC': roc_auc_score(y_test, gcrnn_preds),
            'F1-Score': f1_score(y_test, (gcrnn_preds > 0.5).astype(int)),
            'Accuracy': accuracy_score(y_test, (gcrnn_preds > 0.5).astype(int))
        }

    return results

def print_comparison_table(results):
    df_results = pd.DataFrame(results).T.round(4)
    print("\n" + "="*80)
    print("جدول مقایسه عملکرد مدل‌های پیش‌بینی ریزش (Churn)")
    print("="*80)
    print(df_results.to_markdown())
    print("="*80 + "\n")
    return df_results

def explain_model_shap(model, X_sample, feature_names):
    explainer = shap.TreeExplainer(model)
    shap_values = explainer.shap_values(X_sample)
    if isinstance(shap_values, list) and len(shap_values) > 1:
        shap_vals = shap_values[1][0]
    else:
        shap_vals = shap_values[0]
    print("\nمقادیر SHAP برای نمونه:")
    for i, name in enumerate(feature_names):
        print(f"   {name}: {shap_vals[i]:.4f}")

    # رسم نمودار SHAP
    shap.summary_plot(shap_values, X_sample, feature_names=feature_names, show=False)
    plt.tight_layout()
    plt.title('SHAP Summary Plot for Feature Importance')
    plt.show()

# ===============================
# توابع بهینه‌سازی
# ===============================
def optimize_hyperparameters_random_search(X_train, y_train):
    print("\n--- در حال بهینه‌سازی هایپرپارامترهای XGBoost با Randomized Search ---")

    param_dist = {
        'n_estimators': [100, 200, 300, 400, 500],
        'max_depth': [3, 5, 7, 10, 15],
        'learning_rate': [0.01, 0.05, 0.1, 0.2]
    }

    model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42, n_jobs=1)
    random_search = RandomizedSearchCV(estimator=model, param_distributions=param_dist, n_iter=20, cv=3, scoring='roc_auc', verbose=1, random_state=42, n_jobs=1)
    random_search.fit(X_train, y_train)

    print(f"بهترین هایپرپارامترهای یافت شده توسط Randomized Search: {random_search.best_params_}")

    best_model = random_search.best_estimator_
    return best_model

def main():
    install_libraries()
    INPUT_DATA_PATH = download_data()

    if not INPUT_DATA_PATH or not os.path.exists(INPUT_DATA_PATH):
        print("فایل داده‌ها پیدا نشد. برنامه متوقف می‌شود.")
        return

    try:
        device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
        print(f"در حال بارگذاری داده‌های خام از: {INPUT_DATA_PATH}")
        df = pd.read_csv(INPUT_DATA_PATH)
        if 'customerID' in df.columns:
            df = df.drop('customerID', axis=1)

        df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
        df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].mean())
        df = df.dropna(how='any')
        label_encoder = LabelEncoder()
        categorical_columns = [
            'gender', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService',
            'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
            'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod', 'Churn'
        ]
        for col in categorical_columns:
            df[col] = label_encoder.fit_transform(df[col])

        # --- نمایش Heatmap ---
        print("\n--- نمایش Heatmap ماتریس همبستگی ---")
        plt.figure(figsize=(15, 12))
        sns.heatmap(df.corr(), annot=False, cmap='coolwarm', fmt='.2f')
        plt.title('Heatmap ماتریس همبستگی ویژگی‌های داده‌ها')
        plt.show()

        target_column = 'Churn'
        required_features = [col for col in df.columns if col != target_column]

        X_values = df[required_features].values.astype(float)
        y_values_churn = df[target_column].values.astype(float)
        y_values_cltv = df['MonthlyCharges'].values.astype(float) * df['tenure'].values.astype(float)

        X_train, X_test, y_train, y_test, y_cltv_train, y_cltv_test = train_test_split(
            X_values, y_values_churn, y_values_cltv, test_size=0.2, random_state=42
        )

        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train)
        X_test_scaled = scaler.transform(X_test)

        models_dict = {}
        input_dim = X_train_scaled.shape[1]
        train_loader = DataLoader(TensorDataset(
            torch.tensor(X_train_scaled, dtype=torch.float32),
            torch.tensor(y_train, dtype=torch.float32)
        ), batch_size=64, shuffle=True)

        # --- آموزش مدل‌های کلاسیک ---
        print("\n--- در حال آموزش مدل‌های کلاسیک یادگیری ماشین ---")
        models_to_train = {
            'lgbm': lgb.LGBMClassifier(random_state=42),
            'xgboost': xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
            'decisiontree': DecisionTreeClassifier(random_state=42),
            'randomforest': RandomForestClassifier(random_state=42),
            'bagging': BaggingClassifier(estimator=DecisionTreeClassifier(), random_state=42),
            'adaboost': AdaBoostClassifier(random_state=42),
            'knn': KNeighborsClassifier(),
            'svm': SVC(probability=True, random_state=42)
        }

        for name, model in models_to_train.items():
            print(f"آموزش مدل {name}...")
            if name in ['svm', 'knn']:
                model.fit(X_train_scaled, y_train)
            else:
                model.fit(X_train_scaled, y_train)
            models_dict[name] = model
        print("آموزش مدل‌های کلاسیک کامل شد.")

        # --- بهینه‌سازی هایپرپارامتر با جستجوی تصادفی ---
        optimized_xgb_model = optimize_hyperparameters_random_search(X_train_scaled, y_train)
        if optimized_xgb_model:
            models_dict['xgboost_optimized'] = optimized_xgb_model

        # SHAP for LightGBM
        print("\nتفسیر SHAP برای نمونه اول با LightGBM:")
        explain_model_shap(models_dict['lgbm'], X_test_scaled[:1], required_features)

        # --- آموزش مدل‌های یادگیری عمیق ---
        print("\n--- در حال آموزش مدل‌های یادگیری عمیق ---")

        resnet_model = ResNetForTabular(input_dim=input_dim).to(device)
        print("آموزش ResNet با RMSProp...")
        train_model(resnet_model, train_loader, get_optimizer(resnet_model, 'RMSProp'), nn.BCELoss(), device=device)
        models_dict['resnet'] = resnet_model

        efficientnet_model = EfficientNetLike(input_dim=input_dim).to(device)
        print("آموزش EfficientNet با AdaDelta...")
        train_model(efficientnet_model, train_loader, get_optimizer(efficientnet_model, 'AdaDelta'), nn.BCELoss(), device=device)
        models_dict['efficientnet'] = efficientnet_model

        cnnlstm_model = CNNLSTM(input_dim=input_dim).to(device)
        print("آموزش CNNLSTM...")
        train_model(cnnlstm_model, train_loader, get_optimizer(cnnlstm_model, 'Adam'), nn.BCELoss(), device=device)
        models_dict['cnnlstm'] = cnnlstm_model

        xception_model = XceptionLike(input_dim=input_dim).to(device)
        print("آموزش Xception...")
        train_model(xception_model, train_loader, get_optimizer(xception_model, 'SGD', lr=0.01), nn.BCELoss(), device=device)
        models_dict['xception'] = xception_model

        rnn_model = RNNModel(input_dim=input_dim).to(device)
        print("آموزش RNN...")
        train_model(rnn_model, train_loader, get_optimizer(rnn_model, 'Adam'), nn.BCELoss(), device=device)
        models_dict['rnn'] = rnn_model

        bilstm_cnn_model = BiLSTMCNN(input_dim=input_dim).to(device)
        print("آموزش BiLSTM-CNN...")
        train_model(bilstm_cnn_model, train_loader, get_optimizer(bilstm_cnn_model, 'Adam'), nn.BCELoss(), device=device)
        models_dict['bilstm-cnn'] = bilstm_cnn_model

        print("\nدر حال آموزش GNN و GCRNN...")
        try:
            from torch_geometric.data import Data
            from torch_geometric.nn import GCNConv
            distances = euclidean_distances(X_train_scaled, X_train_scaled)
            edge_indices = np.where(distances < 2.0)
            edge_index = torch.tensor([edge_indices[0], edge_indices[1]], dtype=torch.long).to(device)
            graph_data = Data(x=torch.tensor(X_train_scaled, dtype=torch.float32).to(device), edge_index=edge_index)

            gnn_model = GNNModel(input_dim=input_dim).to(device)
            gnn_optimizer = get_optimizer(gnn_model, 'Adam')
            gnn_model.train()
            for epoch in range(50):
                gnn_optimizer.zero_grad()
                out = gnn_model(graph_data)
                loss = nn.BCELoss()(out.squeeze(), torch.tensor(y_train, dtype=torch.float32).to(device))
                loss.backward()
                gnn_optimizer.step()
            models_dict['gnn'] = gnn_model
            print("آموزش GNN کامل شد.")

            gcrnn_model = GCRNN(input_dim=input_dim).to(device)
            gcrnn_optimizer = get_optimizer(gcrnn_model, 'Adam')
            gcrnn_model.train()
            for epoch in range(50):
                gcrnn_optimizer.zero_grad()
                out = gcrnn_model(graph_data)
                loss = nn.BCELoss()(out.squeeze(), torch.tensor(y_train, dtype=torch.float32).to(device))
                loss.backward()
                gcrnn_optimizer.step()
            models_dict['gcrnn'] = gcrnn_model
            print("آموزش GCRNN کامل شد.")

        except Exception as e:
            print(f"خطا در آموزش مدل‌های GNN/GCRNN: {e}")
            models_dict['gnn'] = None
            models_dict['gcrnn'] = None

        print("\nدر حال آموزش DDQN...")
        state_dim = input_dim
        action_dim = 2
        ddqn_agent = DDQNAgent(state_dim, action_dim, device=device)
        ddqn_env = Environment(states=X_train_scaled, rewards=y_train)
        train_ddqn(ddqn_agent, ddqn_env)
        models_dict['ddqn'] = ddqn_agent

        print("\nدر حال آموزش Shared-Input MTL...")
        model_mtl = SharedMTLModel(input_dim).to(device)
        optimizer_mtl = get_optimizer(model_mtl, 'Adam', lr=0.001)
        criterion_churn = nn.BCELoss()
        criterion_cltv = nn.MSELoss()
        mtl_train_dataset = TensorDataset(
            torch.tensor(X_train_scaled, dtype=torch.float32),
            torch.tensor(y_train, dtype=torch.float32),
            torch.tensor(y_cltv_train, dtype=torch.float32)
        )
        mtl_train_loader = DataLoader(mtl_train_dataset, batch_size=64, shuffle=True)
        train_mtl_model(model_mtl, mtl_train_loader, optimizer_mtl, criterion_churn, criterion_cltv, epochs=50, device=device)
        models_dict['mtl'] = model_mtl
        print("آموزش Shared-Input MTL کامل شد.")

        # -----------------------------------
        # ارزیابی و نتیجه‌گیری
        # -----------------------------------
        print("\n\n--- ارزیابی عملکرد مدل‌ها روی داده‌های تست ---")
        comparison_results = evaluate_all_models(models_dict, X_test, y_test, scaler, device)
        df_results = print_comparison_table(comparison_results)

        # -----------------------------------
        # رسم نمودار مقایسه
        # -----------------------------------
        print("\n--- رسم نمودارهای مقایسه عملکرد مدل‌ها ---")
        fig, axes = plt.subplots(1, 3, figsize=(20, 6))

        # نمودار AUC
        sns.barplot(data=df_results.reset_index(), x='index', y='AUC', ax=axes[0], hue='index', palette='viridis', legend=False)
        axes[0].set_title('مقایسه امتیاز AUC')
        axes[0].set_ylabel('امتیاز AUC')
        axes[0].set_xlabel('نام مدل')
        axes[0].tick_params(axis='x', rotation=45)

        # نمودار F1-Score
        sns.barplot(data=df_results.reset_index(), x='index', y='F1-Score', ax=axes[1], hue='index', palette='plasma', legend=False)
        axes[1].set_title('مقایسه امتیاز F1-Score')
        axes[1].set_ylabel('امتیاز F1-Score')
        axes[1].set_xlabel('نام مدل')
        axes[1].tick_params(axis='x', rotation=45)

        # نمودار Accuracy
        sns.barplot(data=df_results.reset_index(), x='index', y='Accuracy', ax=axes[2], hue='index', palette='magma', legend=False)
        axes[2].set_title('مقایسه دقت')
        axes[2].set_ylabel('دقت')
        axes[2].set_xlabel('نام مدل')
        axes[2].tick_params(axis='x', rotation=45)

        plt.tight_layout()
        plt.show()

        # -----------------------------------
        # نتیجه‌گیری نهایی
        # -----------------------------------
        print("\n" + "="*80)
        print("تحلیل و نتیجه‌گیری نهایی بر اساس عملکرد مدل‌ها:")
        print("="*80)

        df_results = pd.DataFrame(comparison_results).T.sort_values(by='AUC', ascending=False)
        best_model = df_results.index[0]
        best_auc = df_results['AUC'].max()

        print(f"✔️ بهترین مدل بر اساس معیار AUC: **{best_model}** با امتیاز {best_auc:.4f}")
        print("\n**تحلیل مدل‌های مختلف:**")
        print("- **مدل‌های Boosting و بهینه‌سازی شده**: مدل‌های XGBoost و LightGBM به طور کلی عملکرد بسیار قوی دارند. بهینه‌سازی هایپرپارامتر با روش‌های جستجوی تصادفی می‌تواند به دقت بالاتر این مدل‌ها کمک کند. انتظار می‌رود **XGBoost-Optimized** در این مقایسه در جایگاه بالایی قرار گیرد.")
        print("- **مدل‌های Ensemble (RandomForest, Bagging)**: این مدل‌ها نیز به دلیل کاهش واریانس و بهبود پایداری، نتایج خوبی ارائه می‌دهند و به عنوان معیارهای قوی قابل استفاده هستند.")
        print("- **مدل‌های یادگیری عمیق (GNN, GCRNN, CNNLSTM, ResNet)**: این مدل‌ها با توانایی استخراج ویژگی‌های خودکار، پتانسیل کشف الگوهای پیچیده‌تر را دارند. عملکرد آن‌ها به میزان بهینه‌سازی و ساختار داده‌ها وابسته است. GNN و GCRNN به خصوص در داده‌های دارای روابط پیچیده، می‌توانند برتری پیدا کنند.")
        print("- **مدل‌های پایه (DecisionTree, KNN, SVM)**: این مدل‌ها معیارهای اولیه برای مقایسه هستند و نشان می‌دهند که یک راه‌حل ساده تا چه حد می‌تواند مؤثر باشد.")

        print("\n" + "="*80)
        print("تحلیل نهایی کامل شد. می‌توانید کد را با مدل‌ها یا داده‌های دیگر مجددا اجرا کنید.")
        print("="*80)

    except Exception as e:
        print(f"یک خطای کلی رخ داد: {e}")

if __name__ == "__main__":
    main()
```

```
import os
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
import shap
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import LabelEncoder, StandardScaler
import lightgbm as lgb
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, confusion_matrix
from cryptography.fernet import Fernet
import warnings
from collections import deque
import random
import torch.nn.functional as F
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier, AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
import xgboost as xgb
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import euclidean_distances
from torch_geometric.data import Data
from torch_geometric.nn import GCNConv

# Ignore specific warnings
warnings.filterwarnings("ignore", category=UserWarning, module='keras')
warnings.filterwarnings("ignore", category=UserWarning, module='shap')
warnings.filterwarnings("ignore", category=UserWarning, module='sklearn')
warnings.filterwarnings("ignore", category=UserWarning, module='torch_geometric')

# ==========================================================
# Install required libraries
# ==========================================================
def install_libraries():
    """Installs all necessary Python libraries for the script."""
    print("در حال نصب کتابخانه‌های مورد نیاز...")
    try:
        # Install standard libraries
        os.system('pip install gdown==4.4.0 --quiet')
        os.system('pip install lightgbm==3.3.5 --quiet')
        os.system('pip install scikit-learn==1.2.2 --quiet')
        os.system('pip install shap==0.42.1 --quiet')
        os.system('pip install cryptography==41.0.3 --quiet')
        os.system('pip install xgboost --quiet')
        os.system('pip install seaborn matplotlib --quiet')

        # Check and install torch_geometric and its dependencies
        try:
            import torch_geometric
            print("کتابخانه torch_geometric از قبل نصب شده است.")
        except ImportError:
            print("کتابخانه torch_geometric پیدا نشد. در حال نصب...")
            # This command is for PyTorch 2.0 and CUDA 11.8. Adjust if your environment is different.
            os.system('pip install torch-scatter torch-sparse torch-cluster torch-spline-conv torch-geometric -f https://data.pyg.org/whl/torch-2.0.0+cu118.html --quiet')
            # You might need to restart the runtime after this installation in some environments like Google Colab.
            print("برای اطمینان از بارگذاری کامل کتابخانه، ممکن است نیاز به راه‌اندازی مجدد محیط (Runtime Restart) داشته باشید.")

        print("نصب تمامی کتابخانه‌ها کامل شد.")
    except Exception as e:
        print(f"خطا در نصب کتابخانه‌ها: {e}")

# ==========================================================
# Download data from Google Drive
# ==========================================================
def download_data():
    """Downloads the dataset from Google Drive."""
    try:
        import gdown
        file_id = "1YMO2gog6AGiZhlgLCtgHb8Yr3QPxoZ0Q"
        output_file = "telecocustomers.csv"
        if not os.path.exists(output_file):
            print("در حال دانلود فایل داده‌ها...")
            gdown.download(f"https://drive.google.com/uc?id={file_id}", output_file, quiet=False)
        return output_file
    except Exception as e:
        print(f"خطا در دانلود فایل: {e}. مطمئن شوید gdown نصب است.")
        return None

# ==========================================================
# Neural Network Architectures
# ==========================================================
class ResNetForTabular(nn.Module):
    def __init__(self, input_dim, num_classes=1):
        super(ResNetForTabular, self).__init__()
        self.fc_in = nn.Linear(input_dim, 128)
        self.block1 = nn.Sequential(nn.Linear(128, 128), nn.ReLU(), nn.Linear(128, 128))
        self.block2 = nn.Sequential(nn.Linear(128, 128), nn.ReLU(), nn.Linear(128, 128))
        self.fc_out = nn.Linear(128, num_classes)
        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()
    def forward(self, x):
        x = self.relu(self.fc_in(x))
        x = x + self.block1(x)
        x = x + self.block2(x)
        x = self.fc_out(x)
        return self.sigmoid(x)

class EfficientNetLike(nn.Module):
    def __init__(self, input_dim, num_classes=1):
        super(EfficientNetLike, self).__init__()
        self.layers = nn.Sequential(
            nn.Linear(input_dim, 256), nn.ReLU(),
            nn.Linear(256, 128), nn.ReLU(),
            nn.Linear(128, 64), nn.ReLU(),
            nn.Linear(64, num_classes), nn.Sigmoid()
        )
    def forward(self, x):
        return self.layers(x)

class SharedMTLModel(nn.Module):
    def __init__(self, input_dim):
        super(SharedMTLModel, self).__init__()
        self.shared_layers = nn.Sequential(
            nn.Linear(input_dim, 256), nn.ReLU(), nn.Dropout(0.3),
            nn.Linear(256, 128), nn.ReLU(), nn.Dropout(0.3)
        )
        self.churn_head = nn.Sequential(
            nn.Linear(128, 64), nn.ReLU(), nn.Linear(64, 1), nn.Sigmoid()
        )
        self.cltv_head = nn.Sequential(
            nn.Linear(128, 64), nn.ReLU(), nn.Linear(64, 1)
        )
    def forward(self, x):
        shared_features = self.shared_layers(x)
        churn_output = self.churn_head(shared_features)
        cltv_output = self.cltv_head(shared_features)
        return churn_output, cltv_output

class CNNLSTM(nn.Module):
    def __init__(self, input_dim, num_classes=1):
        super(CNNLSTM, self).__init__()
        self.conv1d = nn.Conv1d(in_channels=1, out_channels=64, kernel_size=3, padding=1)
        self.relu = nn.ReLU()
        self.pool1d = nn.MaxPool1d(kernel_size=2)
        lstm_input_dim = 64 * (input_dim // 2)
        self.lstm = nn.LSTM(input_size=lstm_input_dim, hidden_size=128, batch_first=True)
        self.fc = nn.Linear(128, num_classes)
        self.sigmoid = nn.Sigmoid()
    def forward(self, x):
        x = x.unsqueeze(1)
        x = self.conv1d(x)
        x = self.relu(x)
        x = self.pool1d(x)
        x = x.view(x.size(0), 1, -1)
        _, (hn, _) = self.lstm(x)
        out = self.fc(hn.squeeze(0))
        return self.sigmoid(out)

class XceptionLike(nn.Module):
    def __init__(self, input_dim, num_classes=1):
        super(XceptionLike, self).__init__()
        grid_dim = int(np.ceil(np.sqrt(input_dim)))
        self.input_dim = input_dim
        self.grid_dim = grid_dim
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=32, kernel_size=3, padding=1)
        self.sep_conv1 = nn.Sequential(
            nn.Conv2d(32, 32, kernel_size=3, groups=32, padding=1),
            nn.Conv2d(32, 64, kernel_size=1)
        )
        self.relu = nn.ReLU()
        self.avg_pool = nn.AdaptiveAvgPool2d((1, 1))
        self.fc = nn.Linear(64, num_classes)
        self.sigmoid = nn.Sigmoid()
    def forward(self, x):
        x_padded = torch.zeros(x.size(0), self.grid_dim * self.grid_dim).to(x.device)
        x_padded[:, :self.input_dim] = x
        x_2d = x_padded.view(-1, 1, self.grid_dim, self.grid_dim)
        out = self.relu(self.conv1(x_2d))
        out = self.relu(self.sep_conv1(out))
        out = self.avg_pool(out)
        out = out.view(out.size(0), -1)
        out = self.fc(out)
        return self.sigmoid(out)

class GNNModel(nn.Module):
    def __init__(self, input_dim, num_classes=1):
        super(GNNModel, self).__init__()
        self.conv1 = GCNConv(input_dim, 128)
        self.conv2 = GCNConv(128, 64)
        self.fc = nn.Linear(64, num_classes)
    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, p=0.5, training=self.training)
        x = self.conv2(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, p=0.5, training=self.training)
        return torch.sigmoid(self.fc(x))

class GCRNN(nn.Module):
    def __init__(self, input_dim, num_classes=1):
        super(GCRNN, self).__init__()
        self.conv1 = GCNConv(input_dim, 128)
        self.rnn = nn.GRU(128, 64, batch_first=True)
        self.fc = nn.Linear(64, num_classes)
    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = x.unsqueeze(0)
        x, _ = self.rnn(x)
        x = x.squeeze(0)
        return torch.sigmoid(self.fc(x))

class RNNModel(nn.Module):
    def __init__(self, input_dim, hidden_dim=64, num_layers=1, num_classes=1):
        super(RNNModel, self).__init__()
        self.rnn = nn.RNN(input_dim, hidden_dim, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_dim, num_classes)
    def forward(self, x):
        x = x.unsqueeze(1)
        _, hn = self.rnn(x)
        out = self.fc(hn.squeeze(0))
        return torch.sigmoid(out)

class BiLSTMCNN(nn.Module):
    def __init__(self, input_dim, hidden_dim=64, num_layers=1, num_classes=1):
        super(BiLSTMCNN, self).__init__()
        self.conv1d = nn.Conv1d(in_channels=1, out_channels=64, kernel_size=3, padding=1)
        self.pool1d = nn.MaxPool1d(kernel_size=2)
        cnn_output_dim = 64 * (input_dim // 2)
        self.bilstm = nn.LSTM(cnn_output_dim, hidden_dim, num_layers, batch_first=True, bidirectional=True)
        self.fc = nn.Linear(hidden_dim * 2, num_classes)
    def forward(self, x):
        x = x.unsqueeze(1)
        x = self.conv1d(x)
        x = F.relu(x)
        x = self.pool1d(x)
        x = x.view(x.size(0), 1, -1)
        _, (hn, _) = self.bilstm(x)
        hn = torch.cat((hn[-2,:,:], hn[-1,:,:]), dim=1)
        out = self.fc(hn)
        return torch.sigmoid(out)

class DDQN(nn.Module):
    def __init__(self, state_dim, action_dim):
        super(DDQN, self).__init__()
        self.net = nn.Sequential(
            nn.Linear(state_dim, 128), nn.ReLU(),
            nn.Linear(128, 128), nn.ReLU(),
            nn.Linear(128, action_dim)
        )
    def forward(self, state):
        return self.net(state)

class DDQNAgent:
    def __init__(self, state_dim, action_dim, gamma=0.99, lr=0.001, device='cpu'):
        self.state_dim = state_dim
        self.action_dim = action_dim
        self.gamma = gamma
        self.device = device
        self.memory = deque(maxlen=2000)
        self.policy_net = DDQN(state_dim, action_dim).to(device)
        self.target_net = DDQN(state_dim, action_dim).to(device)
        self.target_net.load_state_dict(self.policy_net.state_dict())
        self.target_net.eval()
        self.optimizer = torch.optim.Adam(self.policy_net.parameters(), lr=lr)
        self.loss_fn = nn.MSELoss()

    def choose_action(self, state, epsilon):
        if random.random() < epsilon:
            return random.randrange(self.action_dim)
        with torch.no_grad():
            q_values = self.policy_net(state)
            return q_values.argmax().item()

    def learn(self, batch_size):
        if len(self.memory) < batch_size:
            return
        transitions = random.sample(self.memory, batch_size)
        states, actions, rewards, next_states, dones = zip(*transitions)
        states = torch.tensor(np.array(states), dtype=torch.float32).to(self.device)
        actions = torch.tensor(actions, dtype=torch.long).to(self.device)
        rewards = torch.tensor(rewards, dtype=torch.float32).to(self.device)
        next_states = torch.tensor(np.array(next_states), dtype=torch.float32).to(self.device)
        dones = torch.tensor(dones, dtype=torch.float32).to(self.device)

        q_values = self.policy_net(states).gather(1, actions.unsqueeze(1)).squeeze(1)
        next_q_values_target = self.target_net(next_states).detach()
        next_q_values_policy = self.policy_net(next_states).detach()
        max_q_values = next_q_values_target.gather(1, next_q_values_policy.argmax(dim=1, keepdim=True)).squeeze(1)
        expected_q_values = rewards + self.gamma * max_q_values * (1 - dones)
        loss = self.loss_fn(q_values, expected_q_values)
        self.optimizer.zero_grad()
        loss.backward()
        self.optimizer.step()

    def update_target_net(self):
        self.target_net.load_state_dict(self.policy_net.state_dict())

class Environment:
    def __init__(self, states, rewards):
        self.states = states
        self.rewards = rewards
        self.current_state_idx = 0
    def reset(self):
        self.current_state_idx = random.randrange(len(self.states))
        return self.states[self.current_state_idx]
    def step(self, action):
        reward = self.rewards[self.current_state_idx]
        done = True
        return self.states[self.current_state_idx], reward, done

# ==========================================================
# Training and Evaluation Functions
# ==========================================================
def get_optimizer(model, optimizer_name, lr=0.001):
    if optimizer_name.lower() == 'adam':
        return torch.optim.Adam(model.parameters(), lr=lr)
    elif optimizer_name.lower() == 'sgd':
        return torch.optim.SGD(model.parameters(), lr=lr, momentum=0.9)
    elif optimizer_name.lower() == 'rmsprop':
        return torch.optim.RMSprop(model.parameters(), lr=lr)
    elif optimizer_name.lower() == 'adadelta':
        return torch.optim.Adadelta(model.parameters(), lr=1.0)
    else:
        raise ValueError(f"Optimizer '{optimizer_name}' is not supported.")

def train_model(model, train_loader, optimizer, criterion, epochs=50, device='cpu'):
    model.train()
    for epoch in range(epochs):
        total_loss = 0
        for inputs, targets in train_loader:
            inputs, targets = inputs.to(device), targets.to(device)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, targets.unsqueeze(1).float())
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
        if (epoch + 1) % 10 == 0:
            print(f'Epoch [{epoch+1}/{epochs}], Loss: {total_loss / len(train_loader):.4f}')

def train_mtl_model(model, train_loader, optimizer, criterion_churn, criterion_cltv, epochs=50, device='cpu'):
    model.train()
    for epoch in range(epochs):
        total_loss = 0
        for inputs, targets_churn, targets_cltv in train_loader:
            inputs, targets_churn, targets_cltv = inputs.to(device), targets_churn.to(device), targets_cltv.to(device)
            optimizer.zero_grad()
            churn_preds, cltv_preds = model(inputs)
            loss_churn = criterion_churn(churn_preds.squeeze(), targets_churn.float())
            loss_cltv = criterion_cltv(cltv_preds.squeeze(), targets_cltv.float())
            combined_loss = loss_churn + loss_cltv
            total_loss += combined_loss.item()
            combined_loss.backward()
            optimizer.step()
        if (epoch+1) % 10 == 0:
            print(f'Epoch [{epoch+1}/{epochs}], Combined Loss: {total_loss / len(train_loader):.4f}')

def train_ddqn(agent, env, num_episodes=50, batch_size=32):
    for episode in range(num_episodes):
        state = env.reset()
        done = False
        total_reward = 0
        epsilon = max(0.01, 0.5 - episode / 25)
        while not done:
            state_tensor = torch.tensor(state, dtype=torch.float32).to(agent.device)
            action = agent.choose_action(state_tensor, epsilon)
            next_state, reward, done = env.step(action)
            agent.memory.append((state, action, reward, next_state, done))
            state = next_state
            total_reward += reward
            if len(agent.memory) > batch_size:
                agent.learn(batch_size)
        if (episode + 1) % 10 == 0:
            agent.update_target_net()
            print(f"Episode {episode+1}/{num_episodes}, Total Reward: {total_reward:.2f}")
    print("آموزش DDQN کامل شد.")

def evaluate_all_models(models_dict, X_test, y_test, scaler, device):
    results = {}
    X_test_scaled = scaler.transform(X_test)
    test_loader = DataLoader(TensorDataset(torch.tensor(X_test_scaled, dtype=torch.float32), torch.tensor(y_test, dtype=torch.float32)), batch_size=64, shuffle=False)

    def evaluate_mtl_model(model, data_loader, device):
        model.eval()
        predictions = []
        with torch.no_grad():
            for inputs, _, _ in data_loader:
                inputs = inputs.to(device)
                churn_preds, _ = model(inputs)
                predictions.extend(churn_preds.cpu().numpy().flatten())
        return np.array(predictions)

    classic_models = {
        'LightGBM': models_dict.get('lgbm'),
        'XGBoost': models_dict.get('xgboost'),
        'DecisionTree': models_dict.get('decisiontree'),
        'RandomForest': models_dict.get('randomforest'),
        'Bagging': models_dict.get('bagging'),
        'AdaBoost': models_dict.get('adaboost'),
        'KNN': models_dict.get('knn'),
        'SVM': models_dict.get('svm'),
        'XGBoost-Optimized': models_dict.get('xgboost_optimized')
    }

    all_preds_binary = {}

    for name, model in classic_models.items():
        if model is None:
            continue
        if name in ['SVM', 'Bagging', 'AdaBoost', 'DecisionTree', 'RandomForest', 'KNN', 'LightGBM', 'XGBoost', 'XGBoost-Optimized']:
            if name == 'SVM':
                preds = model.decision_function(X_test_scaled)
                preds_binary = (preds > 0).astype(int)
            else:
                preds = model.predict_proba(X_test_scaled)[:, 1]
                preds_binary = (preds > 0.5).astype(int)
            results[name] = {
                'AUC': roc_auc_score(y_test, preds),
                'F1-Score': f1_score(y_test, preds_binary),
                'Accuracy': accuracy_score(y_test, preds_binary)
            }
            all_preds_binary[name] = preds_binary

    for name, model in [('ResNet', models_dict.get('resnet')),
                       ('EfficientNet', models_dict.get('efficientnet')),
                       ('CNNLSTM', models_dict.get('cnnlstm')),
                       ('Xception', models_dict.get('xception')),
                       ('RNN', models_dict.get('rnn')),
                       ('BiLSTM-CNN', models_dict.get('bilstm-cnn'))]:
        if model is None:
            continue
        model.eval()
        with torch.no_grad():
            preds = model(torch.tensor(X_test_scaled, dtype=torch.float32).to(device)).cpu().numpy().flatten()
            preds_binary = (preds > 0.5).astype(int)
            results[name] = {
                'AUC': roc_auc_score(y_test, preds),
                'F1-Score': f1_score(y_test, preds_binary),
                'Accuracy': accuracy_score(y_test, preds_binary)
            }
            all_preds_binary[name] = preds_binary

    if models_dict.get('mtl'):
        mtl_test_loader = DataLoader(TensorDataset(
            torch.tensor(X_test_scaled, dtype=torch.float32),
            torch.tensor(y_test, dtype=torch.float32),
            torch.tensor(np.zeros_like(y_test), dtype=torch.float32)
        ), batch_size=64, shuffle=False)
        mtl_churn_preds = evaluate_mtl_model(models_dict['mtl'], mtl_test_loader, device=device)
        preds_binary = (mtl_churn_preds > 0.5).astype(int)
        results['Shared-MTL (Churn)'] = {
            'AUC': roc_auc_score(y_test, mtl_churn_preds),
            'F1-Score': f1_score(y_test, preds_binary),
            'Accuracy': accuracy_score(y_test, preds_binary)
        }
        all_preds_binary['Shared-MTL (Churn)'] = preds_binary

    if models_dict.get('gnn') and models_dict.get('gcrnn'):
        adj_matrix = euclidean_distances(X_test_scaled, X_test_scaled)
        edge_indices = np.where(adj_matrix < 2.0)
        edge_index = torch.tensor([edge_indices[0], edge_indices[1]], dtype=torch.long).to(device)
        graph_data_test = Data(x=torch.tensor(X_test_scaled, dtype=torch.float32).to(device), edge_index=edge_index)

        gnn_preds = models_dict['gnn'](graph_data_test).cpu().detach().numpy().flatten()
        preds_binary = (gnn_preds > 0.5).astype(int)
        results['GNN'] = {
            'AUC': roc_auc_score(y_test, gnn_preds),
            'F1-Score': f1_score(y_test, preds_binary),
            'Accuracy': accuracy_score(y_test, preds_binary)
        }
        all_preds_binary['GNN'] = preds_binary

        gcrnn_preds = models_dict['gcrnn'](graph_data_test).cpu().detach().numpy().flatten()
        preds_binary = (gcrnn_preds > 0.5).astype(int)
        results['GCRNN'] = {
            'AUC': roc_auc_score(y_test, gcrnn_preds),
            'F1-Score': f1_score(y_test, preds_binary),
            'Accuracy': accuracy_score(y_test, preds_binary)
        }
        all_preds_binary['GCRNN'] = preds_binary

    return results, all_preds_binary

def print_comparison_table(results):
    df_results = pd.DataFrame(results).T.round(4)
    print("\n" + "="*80)
    print("جدول مقایسه عملکرد مدل‌های پیش‌بینی ریزش (Churn)")
    print("="*80)
    print(df_results.to_markdown())
    print("="*80 + "\n")
    return df_results

def explain_model_shap(model, X_sample, feature_names):
    explainer = shap.TreeExplainer(model)
    shap_values = explainer.shap_values(X_sample)
    if isinstance(shap_values, list) and len(shap_values) > 1:
        shap_vals = shap_values[1][0]
    else:
        shap_vals = shap_values[0]
    print("\nمقادیر SHAP برای نمونه:")
    for i, name in enumerate(feature_names):
        print(f"   {name}: {shap_vals[i]:.4f}")

    shap.summary_plot(shap_values, X_sample, feature_names=feature_names, show=False)
    plt.tight_layout()
    plt.title('SHAP Summary Plot for Feature Importance')
    plt.show()

def plot_confusion_matrix(y_true, y_pred, model_name, ax):
    cm = confusion_matrix(y_true, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax)
    ax.set_title(f'Confusion Matrix for {model_name}')
    ax.set_xlabel('Predicted Label')
    ax.set_ylabel('True Label')
    ax.set_xticklabels(['No Churn', 'Churn'])
    ax.set_yticklabels(['No Churn', 'Churn'])

# ==========================================================
# Hyperparameter Optimization
# ==========================================================
def optimize_hyperparameters_random_search(X_train, y_train):
    print("\n--- در حال بهینه‌سازی هایپرپارامترهای XGBoost با Randomized Search ---")
    param_dist = {
        'n_estimators': [100, 200, 300, 400, 500],
        'max_depth': [3, 5, 7, 10, 15],
        'learning_rate': [0.01, 0.05, 0.1, 0.2]
    }
    model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42, n_jobs=1)
    random_search = RandomizedSearchCV(estimator=model, param_distributions=param_dist, n_iter=20, cv=3, scoring='roc_auc', verbose=1, random_state=42, n_jobs=1)
    random_search.fit(X_train, y_train)
    print(f"بهترین هایپرپارامترهای یافت شده توسط Randomized Search: {random_search.best_params_}")
    best_model = random_search.best_estimator_
    return best_model

# ==========================================================
# Main execution block
# ==========================================================
def main():
    install_libraries()
    INPUT_DATA_PATH = download_data()
    if not INPUT_DATA_PATH or not os.path.exists(INPUT_DATA_PATH):
        print("فایل داده‌ها پیدا نشد. برنامه متوقف می‌شود.")
        return

    try:
        device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
        print(f"در حال بارگذاری داده‌های خام از: {INPUT_DATA_PATH}")
        df = pd.read_csv(INPUT_DATA_PATH)
        if 'customerID' in df.columns:
            df = df.drop('customerID', axis=1)

        df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
        df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].mean())
        df = df.dropna(how='any')
        label_encoder = LabelEncoder()
        categorical_columns = [
            'gender', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService',
            'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
            'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod', 'Churn'
        ]
        for col in categorical_columns:
            df[col] = label_encoder.fit_transform(df[col])

        print("\n--- نمایش Heatmap ماتریس همبستگی ---")
        plt.figure(figsize=(15, 12))
        sns.heatmap(df.corr(), annot=False, cmap='coolwarm', fmt='.2f')
        plt.title('Heatmap ماتریس همبستگی ویژگی‌های داده‌ها')
        plt.show()

        target_column = 'Churn'
        required_features = [col for col in df.columns if col != target_column]
        X_values = df[required_features].values.astype(float)
        y_values_churn = df[target_column].values.astype(float)
        y_values_cltv = df['MonthlyCharges'].values.astype(float) * df['tenure'].values.astype(float)

        X_train, X_test, y_train, y_test, y_cltv_train, y_cltv_test = train_test_split(
            X_values, y_values_churn, y_values_cltv, test_size=0.2, random_state=42
        )

        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train)
        X_test_scaled = scaler.transform(X_test)

        models_dict = {}
        input_dim = X_train_scaled.shape[1]
        train_loader = DataLoader(TensorDataset(
            torch.tensor(X_train_scaled, dtype=torch.float32),
            torch.tensor(y_train, dtype=torch.float32)
        ), batch_size=64, shuffle=True)

        print("\n--- در حال آموزش مدل‌های کلاسیک یادگیری ماشین ---")
        models_to_train = {
            'lgbm': lgb.LGBMClassifier(random_state=42),
            'xgboost': xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
            'decisiontree': DecisionTreeClassifier(random_state=42),
            'randomforest': RandomForestClassifier(random_state=42),
            'bagging': BaggingClassifier(estimator=DecisionTreeClassifier(), random_state=42),
            'adaboost': AdaBoostClassifier(random_state=42),
            'knn': KNeighborsClassifier(),
            'svm': SVC(probability=True, random_state=42)
        }

        for name, model in models_to_train.items():
            print(f"آموزش مدل {name}...")
            model.fit(X_train_scaled, y_train)
            models_dict[name] = model
        print("آموزش مدل‌های کلاسیک کامل شد.")

        optimized_xgb_model = optimize_hyperparameters_random_search(X_train_scaled, y_train)
        if optimized_xgb_model:
            models_dict['xgboost_optimized'] = optimized_xgb_model

        print("\nتفسیر SHAP برای نمونه اول با LightGBM:")
        explain_model_shap(models_dict['lgbm'], X_test_scaled[:1], required_features)

        print("\n--- در حال آموزش مدل‌های یادگیری عمیق ---")

        resnet_model = ResNetForTabular(input_dim=input_dim).to(device)
        print("آموزش ResNet با RMSProp...")
        train_model(resnet_model, train_loader, get_optimizer(resnet_model, 'RMSProp'), nn.BCELoss(), device=device)
        models_dict['resnet'] = resnet_model

        efficientnet_model = EfficientNetLike(input_dim=input_dim).to(device)
        print("آموزش EfficientNet با AdaDelta...")
        train_model(efficientnet_model, train_loader, get_optimizer(efficientnet_model, 'AdaDelta'), nn.BCELoss(), device=device)
        models_dict['efficientnet'] = efficientnet_model

        cnnlstm_model = CNNLSTM(input_dim=input_dim).to(device)
        print("آموزش CNNLSTM...")
        train_model(cnnlstm_model, train_loader, get_optimizer(cnnlstm_model, 'Adam'), nn.BCELoss(), device=device)
        models_dict['cnnlstm'] = cnnlstm_model

        xception_model = XceptionLike(input_dim=input_dim).to(device)
        print("آموزش Xception...")
        train_model(xception_model, train_loader, get_optimizer(xception_model, 'SGD', lr=0.01), nn.BCELoss(), device=device)
        models_dict['xception'] = xception_model

        rnn_model = RNNModel(input_dim=input_dim).to(device)
        print("آموزش RNN...")
        train_model(rnn_model, train_loader, get_optimizer(rnn_model, 'Adam'), nn.BCELoss(), device=device)
        models_dict['rnn'] = rnn_model

        bilstm_cnn_model = BiLSTMCNN(input_dim=input_dim).to(device)
        print("آموزش BiLSTM-CNN...")
        train_model(bilstm_cnn_model, train_loader, get_optimizer(bilstm_cnn_model, 'Adam'), nn.BCELoss(), device=device)
        models_dict['bilstm-cnn'] = bilstm_cnn_model

        print("\nدر حال آموزش GNN و GCRNN...")
        try:
            distances = euclidean_distances(X_train_scaled, X_train_scaled)
            edge_indices = np.where(distances < 2.0)
            edge_index = torch.tensor([edge_indices[0], edge_indices[1]], dtype=torch.long).to(device)
            graph_data = Data(x=torch.tensor(X_train_scaled, dtype=torch.float32).to(device), edge_index=edge_index)

            gnn_model = GNNModel(input_dim=input_dim).to(device)
            gnn_optimizer = get_optimizer(gnn_model, 'Adam')
            gnn_model.train()
            for epoch in range(50):
                gnn_optimizer.zero_grad()
                out = gnn_model(graph_data)
                loss = nn.BCELoss()(out.squeeze(), torch.tensor(y_train, dtype=torch.float32).to(device))
                loss.backward()
                gnn_optimizer.step()
            models_dict['gnn'] = gnn_model
            print("آموزش GNN کامل شد.")

            gcrnn_model = GCRNN(input_dim=input_dim).to(device)
            gcrnn_optimizer = get_optimizer(gcrnn_model, 'Adam')
            gcrnn_model.train()
            for epoch in range(50):
                gcrnn_optimizer.zero_grad()
                out = gcrnn_model(graph_data)
                loss = nn.BCELoss()(out.squeeze(), torch.tensor(y_train, dtype=torch.float32).to(device))
                loss.backward()
                gcrnn_optimizer.step()
            models_dict['gcrnn'] = gcrnn_model
            print("آموزش GCRNN کامل شد.")

        except Exception as e:
            print(f"خطا در آموزش مدل‌های GNN/GCRNN: {e}")
            models_dict['gnn'] = None
            models_dict['gcrnn'] = None

        print("\nدر حال آموزش DDQN...")
        state_dim = input_dim
        action_dim = 2
        ddqn_agent = DDQNAgent(state_dim, action_dim, device=device)
        ddqn_env = Environment(states=X_train_scaled, rewards=y_train)
        train_ddqn(ddqn_agent, ddqn_env)
        models_dict['ddqn'] = ddqn_agent

        print("\nدر حال آموزش Shared-Input MTL...")
        model_mtl = SharedMTLModel(input_dim).to(device)
        optimizer_mtl = get_optimizer(model_mtl, 'Adam', lr=0.001)
        criterion_churn = nn.BCELoss()
        criterion_cltv = nn.MSELoss()
        mtl_train_dataset = TensorDataset(
            torch.tensor(X_train_scaled, dtype=torch.float32),
            torch.tensor(y_train, dtype=torch.float32),
            torch.tensor(y_cltv_train, dtype=torch.float32)
        )
        mtl_train_loader = DataLoader(mtl_train_dataset, batch_size=64, shuffle=True)
        train_mtl_model(model_mtl, mtl_train_loader, optimizer_mtl, criterion_churn, criterion_cltv, epochs=50, device=device)
        models_dict['mtl'] = model_mtl
        print("آموزش Shared-Input MTL کامل شد.")

        print("\n\n--- ارزیابی عملکرد مدل‌ها روی داده‌های تست ---")
        comparison_results, all_preds_binary = evaluate_all_models(models_dict, X_test, y_test, scaler, device)
        df_results = print_comparison_table(comparison_results)

        print("\n--- رسم نمودارهای مقایسه عملکرد مدل‌ها ---")
        fig, axes = plt.subplots(1, 3, figsize=(20, 6))

        sns.barplot(data=df_results.reset_index(), x='index', y='AUC', ax=axes[0], hue='index', palette='viridis', legend=False)
        axes[0].set_title('مقایسه امتیاز AUC')
        axes[0].set_ylabel('امتیاز AUC')
        axes[0].set_xlabel('نام مدل')
        axes[0].tick_params(axis='x', rotation=45)

        sns.barplot(data=df_results.reset_index(), x='index', y='F1-Score', ax=axes[1], hue='index', palette='plasma', legend=False)
        axes[1].set_title('مقایسه امتیاز F1-Score')
        axes[1].set_ylabel('امتیاز F1-Score')
        axes[1].set_xlabel('نام مدل')
        axes[1].tick_params(axis='x', rotation=45)

        sns.barplot(data=df_results.reset_index(), x='index', y='Accuracy', ax=axes[2], hue='index', palette='magma', legend=False)
        axes[2].set_title('مقایسه دقت')
        axes[2].set_ylabel('دقت')
        axes[2].set_xlabel('نام مدل')
        axes[2].tick_params(axis='x', rotation=45)

        plt.tight_layout()
        plt.show()

        print("\n--- رسم ماتریس درهم‌ریختگی برای مدل‌های برتر ---")
        top_models_for_cm = df_results.sort_values(by='AUC', ascending=False).index[:3]
        fig_cm, axes_cm = plt.subplots(1, 3, figsize=(18, 5))

        for i, model_name in enumerate(top_models_for_cm):
            if model_name in all_preds_binary:
                plot_confusion_matrix(y_test, all_preds_binary[model_name], model_name, axes_cm[i])

        plt.tight_layout()
        plt.show()

        print("\n\n================================================================================")
        print("تحلیل نهایی کامل شد. می‌توانید کد را با مدل‌ها یا داده‌های دیگر مجددا اجرا کنید.")
        print("================================================================================")

    except FileNotFoundError:
        print(f"خطا: فایل داده‌ها در مسیر {INPUT_DATA_PATH} پیدا نشد.")
    except Exception as e:
        print(f"یک خطای غیرمنتظره رخ داد: {e}")

if __name__ == '__main__':
    main()
```
